```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('diemk18.csv')
pd.to_numeric(df['tin tich luy'],downcast = 'integer')
pd.to_numeric(df['diem'],downcast = 'float')
excellent_quantity = df[(df['diem']>=3.6)&(df['diem']<=4)].shape
Verygood_quantity = df[(df['diem']>=3.2)&(df['diem']<3.6)].shape
good_quantity = df[(df['diem']>=2.5)&(df['diem']<3.2)].shape
average_quantity = df[(df['diem']>=2)&(df['diem']<2.5)].shape
averageBelow_quantity = df[df['diem']<2].shape
Dict = {'Xut sc':excellent_quantity[0],'Gii':Verygood_quantity[0],'Kh':good_quantity[0],'Trung bnh':average_quantity[0],'Yu':averageBelow_quantity[0]}
#filter dictionary
def filter_Dict(D):
    newD = {}
    for key,value in D.items():
        if value != 0:
            newD[key] = value
    return newD
newDict = filter_Dict(Dict)
#bieu do hoc luc
plt.pie(newDict.values() ,labels= newDict.keys(),autopct='%1.1f%%',shadow = True,startangle=90)
plt.axis('equal')
plt.show()
more_than_134 = df[(df['tin tich luy']>=134)&(df['loai tot nghiep']!='Chua xac dinh')].shape
less_than_134= df[df['tin tich luy']<134].shape
other = df[(df['tin tich luy']>=134)&(df['loai tot nghiep']=='Chua xac dinh')].shape
#bieu do ti le co the ra truong
Dict_credit = {'C kh nng ra trng nm 2022':more_than_134[0],'Khng c kh nng s ra trng nm 2022':other[0]+less_than_134[0]}
plt.pie(Dict_credit.values() ,labels= Dict_credit.keys(),autopct='%1.1f%%',shadow = True,startangle=90)
plt.axis('equal')
plt.show()
#bieu do ti le lam khoa luan
khoaluan = df[df['loai tot nghiep'] == 'Khoa luan'].shape
monthaythe = df[df['loai tot nghiep']== 'Mon Thay The'].shape
other = df[df['loai tot nghiep'] == 'Chua xac dinh'].shape
Dict_graduate = {'khoa luan':khoaluan[0],'mon thay the':monthaythe[0],'chua xac dinh':other[0]}
plt.pie(Dict_graduate.values() ,labels= Dict_graduate.keys(),autopct='%1.1f%%',shadow = True,startangle=90)
plt.axis('equal')
plt.show()
```
