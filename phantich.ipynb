```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('diemk18.csv')
pd.to_numeric(df['tin tich luy'],downcast = 'integer')
pd.to_numeric(df['diem'],downcast = 'float')
excellent_quantity = df[(df['diem']>=3.6)&(df['diem']<=4)].shape
Verygood_quantity = df[(df['diem']>=3.2)&(df['diem']<3.6)].shape
good_quantity = df[(df['diem']>=2.5)&(df['diem']<3.2)].shape
average_quantity = df[(df['diem']>=2)&(df['diem']<2.5)].shape
averageBelow_quantity = df[df['diem']<2].shape
Dict = {'Xuất sắc':excellent_quantity[0],'Giỏi':Verygood_quantity[0],'Khá':good_quantity[0],'Trung bình':average_quantity[0],'Yếu':averageBelow_quantity[0]}
#filter dictionary
def filter_Dict(D):
    newD = {}
    for key,value in D.items():
        if value != 0:
            newD[key] = value
    return newD
newDict = filter_Dict(Dict)
plt.pie(newDict.values() ,labels= newDict.keys(),autopct='%1.1f%%',shadow = True,startangle=90)
plt.axis('equal')
plt.show()
more_than_134 = df[df['tin tich luy']>=134].shape
other = df[df['tin tich luy']<134].shape
Dict_credit = {'Có khả năng ra trường năm 2022':more_than_134[0],'Không có khả năng sẽ ra trường năm 2022':other[0]}
plt.pie(Dict_credit.values() ,labels= Dict_credit.keys(),autopct='%1.1f%%',shadow = True,startangle=90)
plt.axis('equal')
plt.show()
```
